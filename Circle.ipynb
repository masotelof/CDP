```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("Original.jpeg")
rows,cols,_ = img.shape
B,G,R = cv2.split(img)
img = cv2.merge([R,G,B])
plt.imshow(img)
plt.show()
```

```python
g = np.array(0.299 * R + 0.587 * G + 0.114 * B, dtype=np.uint8)
g_img = cv2.merge([g,g,g])
plt.imshow(g_img)
plt.show()
```

```python
g_sft = cv2.medianBlur(g_img, 3)
plt.imshow(g_img)
plt.show()
```

```python
g_edg = cv2.Canny(g_sft,50, 50)
plt.imshow(cv2.merge([g_edg,g_edg,g_edg]))
plt.show()
```

```python
(contour, _) = cv2.findContours(g_edg.copy(),cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img,contour,-1,(0,255,0), 3)
plt.imshow(img)
plt.show()
```
